### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

### Ans:
5  commonly used plots in Seaborn along with their typical use cases:

**Scatter Plot:**

**Use:** Scatter plots are used to visualize the relationship between two numerical variables. They are helpful for identifying patterns, trends, and outliers in data.

**Histogram:**

**Use:** Histograms are used to represent the distribution of a single numerical variable. They help you understand the frequency or probability distribution of the data.

**Bar Plot:**

**Use:** Bar plots are suitable for comparing categorical data or displaying the relationship between a categorical variable and a numerical variable. They can be used to visualize counts, averages, or proportions across categories.

**Box Plot:**

**Use:** Box plots, also known as box-and-whisker plots, are used to display the distribution of a numerical variable and highlight the presence of outliers. They provide information about the median, quartiles, and potential outliers in the data.

**Heatmap:**

**Use:** Heatmaps are used to visualize the correlation or relationships between variables in a dataset. They are especially useful when dealing with large datasets, as they allow you to quickly identify patterns and associations.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('fmri')
```

```python
df.head()
```

```python
sns.lineplot(x=df.timepoint,y=df.signal, data =df, hue='event', style = 'region')
plt.title("FMRI Signal Over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event", loc="upper right")
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
df1 = sns.load_dataset('titanic')
```

```python
df1.head()
```

```python
plt.title("Age Over Pclass")
plt.xlabel("Pclass")
plt.ylabel("Age")
sns.boxplot(data = df1, x='pclass', y='age')
plt.show()
```

```python
plt.title("Fare Over Pclass")
plt.xlabel("Pclass")
plt.ylabel("Fare")
sns.boxplot(data = df1, x='pclass', y='fare')
plt.show()
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
df2 = sns.load_dataset('diamonds')
```

```python
df2.head()
```

```python
plt.figure(figsize=(20,12))
plt.title("Frequency of Price")
plt.xlabel("Price")
sns.histplot(data = df2, x = 'price', hue = 'cut')
plt.show()
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
df4 = sns.load_dataset('flights')
```

```python
df4
```

```python
df5 = df4.pivot_table(index="month", columns="year", values="passengers")
```

```python
df5
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data = df5,annot=True, fmt="d", cmap="YlGnBu")
plt.title("Passenger Counts by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")
```
