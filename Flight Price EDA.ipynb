## EDA And Feature Engineering Flight Price Prediction
### FEATURES
The various features of the cleaned dataset are explained below:
1) Airline: The name of the airline company is stored in the airline column. It is a categorical feature having 6 different airlines.
2) Flight: Flight stores information regarding the plane's flight code. It is a categorical feature.
3) Source City: City from which the flight takes off. It is a categorical feature having 6 unique cities.
4) Departure Time: This is a derived categorical feature obtained created by grouping time periods into bins. It stores information about the departure time and have 6 unique time labels.
5) Stops: A categorical feature with 3 distinct values that stores the number of stops between the source and destination cities.
6) Arrival Time: This is a derived categorical feature created by grouping time intervals into bins. It has six distinct time labels and keeps information about the arrival time.
7) Destination City: City where the flight will land. It is a categorical feature having 6 unique cities.
8) Class: A categorical feature that contains information on seat class; it has two distinct values: Business and Economy.
9) Duration: A continuous feature that displays the overall amount of time it takes to travel between cities in hours.
10) Days Left: This is a derived characteristic that is calculated by subtracting the trip date by the booking date.
11) Price: Target variable stores information of the ticket price.

Here we have to perform feature engineering to convert categorical data into numerical one and while doing so, maintain the importance of data as well

We will start with Date_of_Journey column

```python
df['Date_of_Journey'].str.split('/')
```

Here we are converting each row of 'Date_of_Journey' column into a string and then spliting each element of a row wherever '/' is present into multiple elements of a list as we can see above.

Now using slicing we will seperate day, month and year and store them into new columns

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head(2)
```

Note that the values for day, month and year are in string format so we will change them into integer

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

Now we don't need "Date_of_Journey' separately so we will drop it

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head(2)
```

#### Similarly we will do for other features as well

For arrival time

```python
df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
```

```python
df['Arrival_Hour']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arrival_Minutes']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Arrival_Hour']= df['Arrival_Hour'].astype(int)
df['Arrival_Minutes']=df['Arrival_Minutes'].astype(int)
```

```python
df.info()
```

For Departure time

```python
df['Dep_Hour']=df['Dep_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Dep_Min']=df['Dep_Time'].str.split(' ').str[0].str.split(':').str[1]
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Dep_Hour']=df['Dep_Hour'].astype(int)
df['Dep_Min']=df['Dep_Min'].astype(int)
```

```python
df.info()
```

For Duration

```python
df['Duration_Hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_Hour']=df['Duration_Hour'].str.split('m').str[0]
df['Duration_Min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df['Duration_Min']=df['Duration_Min'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df['Duration_Min']=df['Duration_Min'].astype(int)
df['Duration_Hour']=df['Duration_Hour'].astype(int)
```

```python
df.info()
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df.head(2)
```

#### Since Route feature gives info about source and destination which is already present as separate features, so we will drop route

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head()
```

#### Since Total_Stops is a categorical feature, we will use target guided encoding 

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].mode()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df.head()
```

#### For Airline, Source and Destination we will use OHE encoding

```python
df['Airline'].unique()
```

```python
df['Source'].unique()
```

```python
df['Destination'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
```

```python
encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```

```python
df1=pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
df_final=pd.concat([df,df1],axis=1)
```

```python
df_final.head()
```

```python
df_final.drop(['Airline','Source','Destination'],axis=1,inplace=True)
```

```python
df_final.head()
```

#### For Additional_Info we will again use target encoding

```python
df_final['Additional_Info'].unique()
```

```python
df_final['Additional_Info']=df_final['Additional_Info'].map({'No info':0,'In-flight meal not included':1,'No check-in baggage included':2,'1 Long layover':3,'Change airports':4,'Business class':5,'No Info':0,'1 Short layover':6,'Red-eye flight':7,'2 Long layover':8})
```
