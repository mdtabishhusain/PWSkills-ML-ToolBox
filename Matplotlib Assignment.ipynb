### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
Note: Also add title, xlabel, and ylabel to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

### Ans:
Scatter plots display individual data points as dots on a two-dimensional plane, with one variable on the x-axis and another on the y-axis. They are used to visualize the relationship between two continuous variables.

```python
plt.title("Scatter Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.scatter(x=x, y=y)
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

### Ans:
The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. Each subplot is a smaller, individual plot that shares the same figure canvas. This function is valuable for creating multi-panel or multi-axis visualizations in a single figure, allowing you to display multiple plots or charts side by side for easy comparison or presentation

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2= np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')
plt.tight_layout()
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

### Ans:
Bar charts represent data as rectangular bars, with the length of each bar proportional to the value it represents. They are often used to compare discrete categories or groups.

```python
plt.title("Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.bar(x=company,height=profit)
```

```python
plt.title("Horizontal Bar Plot")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.barh(y=company,width=profit)
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

### Ans:
Box plot is a graphical representation of the distribution of a dataset. It is also known as a box-and-whisker plot. Box plots are used to visualize the distribution of a dataset, including its median, quartiles, and potential outliers. 

```python
plt.boxplot([box1 , box2], labels =['Box1', 'Box2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```
