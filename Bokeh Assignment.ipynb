### Q1. How can you create a Bokeh plot using Python code?

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show
```

```python
p = figure(title="Simple Line Plot", x_axis_label="X-axis", y_axis_label="Y-axis")
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
output_file('line.html')
p.line(x, y, line_width=2, line_color="blue", legend_label="Line")
p.legend.title = "Legend"
p.legend.label_text_font_size = "12pt"
show(p)
```

### Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

### Ans:
In Bokeh, "glyphs" are geometric shapes or graphical marks used to represent data on a plot. Glyphs can be used to visualize various types of data, such as points, lines, bars, circles, rectangles, and more. They are the fundamental building blocks for creating visualizations in Bokeh.

To add glyphs to a Bokeh plot, you typically follow these steps:
1. Import Bokeh Modules
2. Create a Figure
3. Define Data
4. Add Glyphs
5. Customize Glyph Properties
6. Display or Save the Plot

```python
p = figure(title="Glyphs Example", x_axis_label="X-axis", y_axis_label="Y-axis")
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
p.circle(x, y, size=10, color="blue", legend_label="Circles")
p.line(x, y, line_width=2, line_color="red", legend_label="Line")
p.legend.title = "Legend"
p.legend.label_text_font_size = "12pt"
show(p)
```

### Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

### ANS:
You can customize the appearance of a Bokeh plot, including the axes, title, and legend, by setting various attributes and properties of the plot and its components

```python
p = figure(
    title="Customized Bokeh Plot",
    x_axis_label="X-axis Label",
    y_axis_label="Y-axis Label",
    plot_width=600,
    plot_height=400,
    background_fill_color="lightgray",
    border_fill_color="white",
)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
p.line(x, y, line_width=2, line_color="blue", legend_label="Line")
p.circle(x, y, size=10, color="red", legend_label="Circles")
p.xaxis.axis_label_text_font_size = "14pt"
p.yaxis.axis_label_text_font_size = "14pt"
p.legend.title = "Legend Title"
p.legend.label_text_font_size = "12pt"
p.legend.location = "top_right"
show(p)
```

### Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

### Ans:
A Bokeh server is a Python process that runs a Bokeh application. It allows you to define and serve interactive Bokeh plots and applications as web services. With a Bokeh server, you can create web-based dashboards, data exploration tools, and dynamic visualizations that can be accessed through a web browser. The Bokeh server handles the communication between the client (web browser) and the server-side Python code, enabling real-time updates and interactivity.
