Q6. You are working on a project to build a model to predict stock prices. The dataset contains many features, such as
company financial data and market trends. Explain how you would use PCA to reduce the dimensionality of the dataset.

Principal Component Analysis (PCA) is a dimensionality reduction technique used to transform a high-dimensional dataset into a lower-dimensional space while retaining as much of the original data's variance as possible. When building a model to predict stock prices, PCA can help in reducing the dimensionality of the dataset, which is crucial for improving model efficiency and reducing the risk of overfitting. Here's how you would use PCA for this purpose:

Understand the Dataset and Features:

Thoroughly understand the dataset and the features it contains, including company financial data and market trends.
Preprocess the Data:

Clean and preprocess the dataset, handling missing values, outliers, and any data quality issues.
Standardize the Data:

Standardize the features to have a mean of 0 and a standard deviation of 1. This is important for PCA, as it is a variance-based technique.
Calculate the Covariance Matrix:

Calculate the covariance matrix of the standardized data, which represents the relationships between different features.
Compute Eigenvectors and Eigenvalues:

Compute the eigenvectors and eigenvalues of the covariance matrix. Eigenvectors represent the directions of maximum variance, and eigenvalues indicate the magnitude of the variance in each direction.
Sort Eigenvalues:

Sort the eigenvalues in descending order. This step is crucial as it helps prioritize the most significant components (those associated with higher eigenvalues).
Select Principal Components (PCs):

Choose the top 'k' eigenvectors based on the explained variance you want to retain. Typically, you select enough principal components to capture a certain percentage (e.g., 95%) of the total variance.
Project Data onto New Feature Space:

Project the original data onto the new feature space formed by the selected principal components. This involves computing the dot product of the original data with the selected eigenvectors.
Use Reduced Dimensions for Modeling:

Use the reduced-dimensional dataset (projected onto the selected principal components) for further modeling tasks, such as predicting stock prices.

Q7. For a dataset containing the following values: [1, 5, 10, 15, 20], perform Min-Max scaling to transform the values to a
range of -1 to 1.

```python
import numpy as np

# Given dataset
dataset = np.array([1, 5, 10, 15, 20])

# Define the desired range (-1 to 1)
a = -1
b = 1

# Calculate min and max of the dataset
min_val = np.min(dataset)
max_val = np.max(dataset)

# Apply Min-Max scaling
scaled_dataset = ((dataset - min_val) / (max_val - min_val)) * (b - a) + a

# Print the scaled dataset
print(scaled_dataset)
```

Q8. For a dataset containing the following features: [height, weight, age, gender, blood pressure], perform Feature
Extraction using PCA. How many principal components would you choose to retain, and why?

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for the features: height, weight, age, blood pressure
# Replace this with your actual standardized data
data = np.array([
    [0.5, 0.7, 0.2, 0.3],
    [0.4, 0.6, 0.1, 0.25],
    [0.45, 0.65, 0.15, 0.28],
    # ... Add more data rows as needed
])

# Calculate the covariance matrix
covariance_matrix = np.cov(data, rowvar=False)

# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Sort eigenvalues in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues_sorted = eigenvalues[sorted_indices]

# Calculate explained variance
explained_variance_ratio = eigenvalues_sorted / np.sum(eigenvalues_sorted)
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Plot explained variance vs. number of components
plt.plot(cumulative_explained_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance vs. Number of Components')
plt.grid()
plt.show()

# Choose the number of components to retain (e.g., where explained variance is 95% or more)
# Example: If explained variance is 95%, choose the corresponding number of components
num_components_to_retain = np.argmax(cumulative_explained_variance >= 0.95) + 1
print("Number of components to retain:", num_components_to_retain)
```

would choose the number of components that retain a high percentage of variance (e.g., 95%) to ensure that you capture most of the relevant information in the dataset. The specific number of components may vary based on the desired level of variance retention and the specific requirements of your analysis or modeling task.
